```python
import pandas as pd

df = pd.read_csv("SR2024.csv", on_bad_lines="skip")
df = df[df["First 3 Chars of Postal Code"] == "M4E"]
df
```

```python
import matplotlib.pyplot as plt

section_counts = df["Section"].value_counts()

plt.figure(figsize=(10,6))
section_counts.plot(kind="barh")

plt.title("Number of Service Requests per Section")
plt.xlabel("Count")
plt.ylabel("Section")
plt.show()
```

```python
sections = df["Section"].unique()

for sec in sections:
    plt.figure(figsize=(10,6))
    subset = df[df["Section"] == sec]
    subset["Service Request Type"].value_counts().plot(kind="barh")
    
    plt.title(f"Service Request Types in Section: {sec}")
    plt.xlabel("Count")
    plt.ylabel("Service Request Type")
    plt.show()
```

```python
collections_values = df.loc[df["Section"] == "Collections", "Service Request Type"].unique()
plt.figure(figsize=(12, 10))  # 增大画布


import matplotlib.pyplot as plt

def plot_section_barplot(df, section="Collections", top_n=None, fontsize=12):
    """
    绘制某个 Section 下 Service Request Type 的分布
    
    参数：
        df : pandas DataFrame
        section : str, 要选择的 Section 名字
        top_n : int or None, 如果指定，就只显示前 top_n 个类型
        fontsize : int, 字体大小
    """
    # 筛选出该 Section 的 Service Request Type
    counts = df.loc[df["Section"] == section, "Service Request Type"].value_counts()
    
    # 如果需要，只保留前 top_n 个
    if top_n is not None:
        counts = counts.head(top_n)
    
    # 画图
    plt.figure(figsize=(12, 8))
    counts.plot(kind="barh")
    
    plt.title(f"Service Request Types in Section: {section}", fontsize=fontsize+4)
    plt.xlabel("Count", fontsize=fontsize)
    plt.ylabel("Service Request Type", fontsize=fontsize)
    plt.xticks(fontsize=fontsize)
    plt.yticks(fontsize=fontsize)
    plt.gca().invert_yaxis()  # 让最多的类别排在最上面
    plt.show()


# 调用示例：展示 Collections 下的前 20 类，字体大小 12
plot_section_barplot(df, section="Collections", top_n=20, fontsize=12)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# 所有类别
categories = df.loc[df["Section"] == "Collections", "Service Request Type"].unique()

# 转换成 TF-IDF 特征
vectorizer = TfidfVectorizer(stop_words="english")
X = vectorizer.fit_transform(categories)

# 聚类，比如设定成 10 类
kmeans = KMeans(n_clusters=5, random_state=0)
df_clusters = pd.DataFrame({
    "Category": categories,
    "Cluster": kmeans.fit_predict(X)
})

print(df_clusters)
```

```python
# Merge cluster info into your original df
df_merged = df.merge(df_clusters, 
                     left_on="Service Request Type", 
                     right_on="Category", 
                     how="left")
cluster_counts = df_merged["Cluster"].value_counts().sort_index()
print(cluster_counts)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
cluster_counts.plot(kind="bar")

plt.title("Distribution of Service Requests by Cluster")
plt.xlabel("Cluster ID")
plt.ylabel("Number of Requests")
plt.show()
```

```python
import numpy as np

# 获取特征名字（即单词）
terms = vectorizer.get_feature_names_out()

# 遍历每个 cluster
num_top_words = 8  # 每个 cluster 显示前8个关键词
for i in range(kmeans.n_clusters):
    # 找到该 cluster 中最重要的若干个词
    center = kmeans.cluster_centers_[i]
    top_indices = center.argsort()[-num_top_words:][::-1]
    top_terms = [terms[ind] for ind in top_indices]
    
    print(f"Cluster {i}: {', '.join(top_terms)}")
```
